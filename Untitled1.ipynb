```python
from datasets import load_dataset

dataset = load_dataset('trivia_qa', name = 'rc')

train = dataset['train']

tmp_data = train.select(range(11))
tmp_data = tmp_data.map(lambda example: {'wiki_text': example['entity_pages']['wiki_context'], 'answer': example['answer']['normalized_value']}, remove_columns=['search_results', 'question_source', 'answer', 'entity_pages'])

tmp_data = tmp_data.filter(lambda example: len(example['wiki_text']) == 1)

tmp_data = tmp_data.map(lambda example: {'wiki_text_lower': example['wiki_text'][0].lower()})
```

```python
import torch
from transformers import BertTokenizer, BertModel, BertForMaskedLM, BertTokenizerFast
```

```python
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')
```

```python
tokenized_text = tokenizer.tokenize(tmp_data[0]["wiki_text"][0])
tokenized_text
```

```python
CLS_token = "CLS"
SEP_token = "SEP"

if tokenized_text[0] != CLS_token:
    tokenized_text.insert(0, CLS_token)


test = ['hej', '(']
for word in range(len(tokenized_text)):
    if (tokenized_text[word].isnumeric() == False and tokenized_text[word].isalpha() == False):
        #print(tokenized_text[word])
        tokenized_text[word] = SEP_token
        #print(tokenized_text[word])
```
